Basic Inspection and Exploration

```python
import pandas as pd
import seaborn as sns

# Load from seaborn or CSV
iris = sns.load_dataset('iris')
# or: iris = pd.read_csv('Iris.csv')

print(iris.head())
print("Shape:", iris.shape)
print(iris.info())
print(iris.describe())

# Check for missing values and duplicates
print("Missing values:\n", iris.isnull().sum())
print("Duplicate rows count:", iris.duplicated().sum())

```

Visualization: Histograms, Box Plots, Scatter, Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for sepal/petal features
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
iris[features].hist(figsize=(10, 8), bins=10)
plt.show()

# Boxplot to visualize distribution and outliers
for feature in features:
    sns.boxplot(x=iris[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()

# Pair plot with species hue
sns.pairplot(iris, hue='species')
plt.show()

# Basic scatter plot of sepal length vs width
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=iris['species'].astype('category').cat.codes)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Width')
plt.show()

```

Correlation and Heatmap

```python
corr = iris[features].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Dimensionality Reduction with PCA (3D Visualization)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.datasets import load_iris
data = load_iris(as_frame=True)
df = data.frame

X = df[data.feature_names]
y = data.target

pca = PCA(n_components=3)
X_reduced = pca.fit_transform(X)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2],
    c=y, s=40, cmap='viridis'
)
ax.set_title('3D PCA of Iris Dataset')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
legend1 = ax.legend(*scatter.legend_elements(), title="Species")
ax.add_artist(legend1)
plt.show()
```

Train/Test Split and Baseline Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = iris_df.drop(columns=['species', 'species_encoded'])
y = iris_df['species_encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)
print("Accuracy:", model.score(X_test, y_test))

```
